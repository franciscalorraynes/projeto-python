```python
kmeans = KMeans(n_clusters=5) #quantidade de grupos
k_fit = kmeans.fit(data)
#ir na biblioteca do kmins 
```

```python
predicoes = k_fit.labels_ #rotulos
data['Grupos'] = predicoes 
```

```python
data.head()
```

```python
data["Grupos"].value_counts()
```

```python
data.groupby('Grupos').mean()
```

```python
col_list=list(data)
ext=col_list[0:10]
est=col_list[10:20]
agr=col_list[20:30]
csn=col_list[30:40]
opn=col_list[40:50]
```

```python
data_soma=pd.DataFrame()
data_soma['Extroversão']=data[ext].sum(axis=1)/10
data_soma['Neurótico'] = data[est].sum(axis=1)/10
data_soma['Agradável'] = data[agr].sum(axis=1)/10
data_soma['Responsável'] = data[csn].sum(axis=1)/10
data_soma['Aberto'] = data[opn].sum(axis=1)/10
data_soma['Grupos'] = predicoes
```

```python
data_soma.groupby('Grupos').mean()
```

```python
data_clusters = data_soma.groupby('Grupos').mean()
```

```python
plt.figure(figsize=(22,3))
for i in range(0, 5):
    plt.subplot(1,5,i+1)
    plt.bar(data_clusters.columns, data_clusters.iloc[:, i], color='green', alpha=0.2)
    plt.plot(data_clusters.columns, data_clusters.iloc[:, i], color='red')
    plt.title('Grupo ' + str(i))
    plt.xticks(rotation=45)
    plt.ylim(0,4);
```
